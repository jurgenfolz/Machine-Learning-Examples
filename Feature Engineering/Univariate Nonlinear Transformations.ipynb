# Univariate Nonlinear Transformations

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#This code is used to generate synthetic data for a regression problem.
rnd = np.random.RandomState(0)
X_org = rnd.normal(size=(1000, 3))
w = rnd.normal(size=3)
X = rnd.poisson(10 * np.exp(X_org))
y = np.dot(X_org, w)
print(f"Number of feature appearances: {np.bincount(X[:, 0])}")
```

```python
# Plot the distribution of the first feature values
bins = np.bincount(X[:, 0])
plt.bar(range(len(bins)), bins, color='red')
plt.ylabel("Number of appearances")
plt.xlabel("Value")
plt.title("Distribution of the 1st feature values")
plt.show()
```

```python
# Plot the distribution of the 2nd feature values
bins = np.bincount(X[:, 1])
plt.bar(range(len(bins)), bins, color='blue')
plt.ylabel("Number of appearances")
plt.xlabel("Value")
plt.title("Distribution of the 2nd feature values")
plt.show()
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
score = Ridge().fit(X_train, y_train).score(X_test, y_test)
print("Test score: {:.3f}".format(score))
```

The low R2 score shows that Ridge was not able to really capture the relationship between X and y
