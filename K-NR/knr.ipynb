# K-Neighbors Regressor

```python
#Imports
from sklearn.neighbors import KNeighborsRegressor
import mglearn
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X, y = mglearn.datasets.make_wave(n_samples=40)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
reg = KNeighborsRegressor(n_neighbors=3)
reg.fit(X_train, y_train)
print("Test set predictions:\n{}".format(reg.predict(X_test)))
print("Test set R^2: {:.2f}".format(reg.score(X_test, y_test)))
```

```python
import matplotlib.pyplot as plt

# Generate a range of feature values
X_range = np.linspace(X.min(), X.max(), 1000).reshape(-1, 1)

# Predict the target values for the generated feature values
y_pred = reg.predict(X_range)

# Plot the training and test data
plt.scatter(X_train, y_train, color='blue', label='Train data')
plt.scatter(X_test, y_test, color='red', label='Test data')

# Plot the predictions as a line
plt.plot(X_range, y_pred, color='green', label='Prediction')

plt.xlabel('Feature')
plt.ylabel('Target')
plt.legend()
plt.show()
```
